```python
def calc_drag_force(speed):
    # estimate aero drag
    density = 1.225 # air est
    c_d = 0.15 # estimate
    width = 0.310 # m - 12.2 in
    height = 0.165 # m - 6.5 in
    A = width * height * 0.8
    f_drag = 0.5 * density * c_d * A * speed**2
    # print("f_drag: ", f_drag, " at ", speed / mph_to_mps)
    return f_drag

def calc_down_force(speed):
    # estimate aero down force
    density = 1.225 # air est
    c_d = 0.15 # estimate
    width = 0.310 # m - 12.2 in
    height = 0.165 # m - 6.5 in
    A = width * height * 0.2
    f_down_force = 0.5 * density * c_d * A * speed**2
    # print("f_drag: ", f_drag, " at ", speed / mph_to_mps)
    return f_down_force
    

amps = range(0, 550, 50)
for amp in amps:
    volts = np.linspace(0, motor.motorParams.max_voltage, 100)
    rpms = np.vectorize(motor.calc_motor_rpm_from_volts)(volts)
    speeds = np.vectorize(driveTrain.calc_car_speed_mps)(rpms)
    motor_torques = np.vectorize(motor.calc_motor_torque)(amp, rpms)
    wheel_torques = np.vectorize(driveTrain.calc_wheel_torque_from_motor_torque)(motor_torques)
    wheel_forces = np.vectorize(driveTrain.calc_wheel_force_from_wheel_torque)(wheel_torques)
    drag_forces = np.vectorize(calc_drag_force)(speeds)
    accelerations = (wheel_forces - drag_forces) / 8.0

    # plt.plot(speeds /  mph_to_mps, accelerations, label=str(amp) + 'A')
    # plt.grid()
    # plt.xlabel('Speed (mph)')
    # plt.ylabel('Acceleration (m/s^2)')
    # plt.legend()

    plt.plot(speeds /  mph_to_mps, wheel_forces, label='F@350A')
    plt.plot(speeds /  mph_to_mps, drag_forces, label='_Drag')
    plt.grid()
    plt.xlabel('Speed (mph)')
    plt.ylabel('Forces (N)')
    plt.legend()
    # plt.show()
```
