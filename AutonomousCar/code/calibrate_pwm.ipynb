```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pwm_calibration_data.csv", names=['set', 'throttle', 'steer', 'check'])

df['throttle error'] = df['set'] - df['throttle']
df['steer error'] = df['set'] - df['steer']

plt.plot(df['set'], df['throttle error'], 'b.')
plt.plot(df['set'], df['steer error'], 'r.')
plt.ylim([-20, 20])
plt.show()

print("mean throttle error: ", df['throttle error'].mean())
print("mean steer error: ", df['steer error'].mean())
```
