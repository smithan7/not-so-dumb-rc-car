```python
import threading
import socket
import numpy as np
import random
import time
```

```python
class RC_Logger:
    '''This has 4 threads
    1 - Monitors the socket connection to the UI
    2 - Interacts with OLA
    3 - Interacts with GPS
    4 - Interacts with PWM
    '''

    def __init__(self):
            print("Initializing RC_Logger")
            self.GPS_thread = threading.Thread(target=self.GPS_fun, daemon=True)
            self.OLA_thread = threading.Thread(target=self.OLA_fun, daemon=True)
            self.PWM_thread = threading.Thread(target=self.PWM_fun, daemon=True)
            # self.Socket_thread = threading.Thread(target=self.Socket_fun, args=(1,), daemon=True)
            
            self.GPS_lock = threading.Lock()
            self.OLA_lock = threading.Lock()
            self.PWM_lock = threading.Lock()
            # self.Socket_lock = threading.Lock()

            self.GPS_has_new_data = False
            self.OLA_has_new_data = False
            self.PWM_has_new_data = False
            self.running = True
            
            self.GPS_thread.start()
            self.OLA_thread.start()
            self.PWM_thread.start()
            # self.Socket_thread.start()
            
            self.Sock_fun()
            

    def GPS_fun(self):
        while self.running:
            self.GPS_lock.acquire()
            self.lat = random.random()
            self.lon = random.random()
            self.alt = random.random()
            self.GPS_has_new_data = True
            self.GPS_lock.release()
            time.sleep(0.1)
        
    def OLA_fun(self):
        while self.running:
            self.OLA_lock.acquire()
            self.ax = random.random()
            self.ay = random.random()
            self.az = random.random()
            self.OLA_has_new_data = True
            self.OLA_lock.release()
            time.sleep(0.1)
        
    def PWM_fun(self):
        while self.running:
            self.PWM_lock.acquire()
            self.steer = random.random()
            self.throttle = random.random()
            self.PWM_has_new_data = True
            self.PWM_lock.release()
            time.sleep(0.1)
        
    def Sock_fun(self):
        print("Entering for loop")
        start_time = time.time()
        cnt = 0
        while time.time() - start_time < 10.0:
            cnt = cnt + 1
            if self.GPS_has_new_data and not self.PWM_lock.locked():
                self.GPS_lock.acquire()
                print("GPS: ", self.lat, ", ", self.lon, " ,", self.alt)
                self.GPS_has_new_data = False
                self.GPS_lock.release()
            
            if self.OLA_has_new_data and not self.OLA_lock.locked():
                self.OLA_lock.acquire()
                print("OLA: ", self.ax, ", ", self.ay, ", ", self.az)
                self.OLA_has_new_data = False
                self.OLA_lock.release()
            
            if self.PWM_has_new_data and not self.PWM_lock.locked():
                self.PWM_lock.acquire()
                print("PWM: ", self.steer, ", ", self.throttle)
                self.PWM_has_new_data = False
                self.PWM_lock.release()
    
        print("cnt: ", cnt)
        
print("A")
rc = RC_Logger()
print("B")
```
