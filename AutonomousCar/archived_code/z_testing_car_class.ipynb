```python
import numpy  as np
import matplotlib.pyplot as plt
from Test_Car_2D import Test_Car_2D
```

```python
"""Test the car by applying some inputs"""
car = Test_Car_2D()

u_steer = 0.0
dt = 0.1
speeds = []
times = []
for i in range(10):
    u_accel = car.linear_speed_dynamics([10.0, 0])
    car.kinematic_update([u_steer, u_accel], dt)
    speeds.append(car.speed)
    times.append(i*dt)

plt.plot(times, speeds)
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.show()
```

```python
"""Test the car by applying some inputs"""
car = Test_Car_2D()

u = [1.0, 0.1]
dt = 0.1
x_speeds = []
y_speeds = []
heading_speeds = []
steering_speeds = []
times = []
for i in range(100):
    car.kinematic_update(u, dt)
    x_speeds.append(car.xd[0])
    y_speeds.append(car.xd[1])
    heading_speeds.append(car.xd[2])
    steering_speeds.append(car.xd[3])
    times.append(i*dt)

plt.plot(times, heading_speeds, 'r', label='Heading Speed')
plt.plot(times, steering_speeds, 'b', label='Steering Speed')
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Speed (rads/s)")
plt.legend()
plt.show()

plt.plot(times, x_speeds, 'r', label='X Speed')
plt.plot(times, y_speeds, 'b', label='Y Speed')
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.legend()
plt.show()
```

```python
"""Test the car by applying some inputs"""
car = Test_Car_2D()
car.speed = 5.0
car.x[3] = 0.3

u_steer = 0.1
dt = 0.1
x = []
y = []
heading = []
steering = []
times = []
for i in range(100):
    u_accel = car.linear_speed_dynamics([0.1, 0])
    car.kinematic_update([u_accel, u_steer], dt)
    x.append(car.x[0])
    y.append(car.x[1])
    heading.append(car.x[2])
    steering.append(car.x[3])
    times.append(i*dt)

plt.plot(times, heading, 'r.', label='Heading Angle')
plt.plot(times, steering, 'b.', label='Steering Angle')
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Angle (rads)")
plt.legend()
plt.show()

plt.plot(times, x, 'r', label='X')
plt.plot(times, y, 'b', label='Y')
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Pos (m)")
plt.legend()
plt.show()

plt.plot(x, y,'r', label='X')
plt.grid()
plt.xlabel("Pos X (m)")
plt.ylabel("Pos Y (m)")
plt.legend()
plt.axis('equal')
plt.show()
```
